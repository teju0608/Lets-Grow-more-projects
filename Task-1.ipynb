#### Data Visulization

Find hot zones using data visulization

```python
plt.figure(figsize = (10,10))
sns.countplot(y = df.Year)
plt.title('Attacks by year')
plt.show()
```

From the above graph, we can see that the more attacks were happened on 2014 and then in 2015.  Here we see that from 1970 onwards, the last 6 years attacks score is more. But from 2014 attacks score is decreasing.

```python
# attacks by month

plt.figure(figsize = (10,5))
sns.countplot(x = df.Month)
plt.title("Attacks by Month")
plt.show()
```

From the above graph we can see that,almost every month had same attacks was happened.

```python
# attacks by day
plt.figure(figsize = (15,5))
sns.countplot(x = df.Day)

plt.title('Attacks by Day')
plt.show()
```

From the above graph we can see that, almost every day had same attacks was happened. At day 31 attacks are low and high during 1st and 15th day.

```python
# attacks by country

plt.figure(figsize = (10,30))
plt.rcParams.update({'font.size':8});
sns.countplot(y = df.Country)
plt.title('Attacks by Country')
plt.show()
```

From the above graph we can see that, Iraq country has the highest number of attacks then Pakistan, then Afganistan, India and then follow it.
Hence government should follow safety in these countries.

```python
# attacks by region

plt.figure(figsize = (20,5))
sns.countplot(x = df.Region)

plt.title('Attacks by Region')
plt.show()
```

From the above graph we can see that, the attacks happened in firstly Middle East and North Africa and secondly South asia.

```python
# target type

plt.figure(figsize = (20,5))
plt.rcParams.update({'font.size':12});
plt.xticks(rotation = 90)
sns.countplot(x = df.Target_Type)

plt.title('Attacks based on target')
plt.show()
```

The attacks based on target is Private citizens and Property, then miliatry.

```python
plt.figure(figsize = (20,5))
plt.rcParams.update({'font.size':12});
plt.xticks(rotation = 90)
sns.countplot(x = df.Attack_Type)

plt.title('Based on attack type')
plt.show()
```

From the above graph we can see that, the most attack types are Bombing or Explosion.
